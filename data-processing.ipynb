## 텍스트를 비교해서 달라진 부분만 뽑아보는 방법

* 원문 판독문과 비교할 각 연구자별 판독문을 엑셀파일에서 뽑아 동일한 기준으로 비교할 수 있게 만든다.
* 이후 '다른 글자'만 자동으로 뽑는다.

결과는 직접 검증한 뒤, 엑셀 파일에 업데이트 함 

````python
# 엑셀 파일에서 연구자별 판독문 데이터 시트(reading_rawdata)를 읽어온다. 

df = pd.read_excel('studies.xlsx', sheet_name='reading_rawdata')
````

````python
df = df.fillna('')  # 편의상 빈 셀을 'NaN'이 아닌 공백으로 보이게 하기 위해서 
df
````

````python
# 경주문화재연구소의 2019 버전(id=1)으로 판독(reading)의 기준으로 삼는다.
# 앞으로 다른 연구의 판독문과 비교하기 위해 
# id=1 의 판독(reading)을 뽑아 'base_text' 라는 변수에 넣어둔다.
# 중성리비는 12줄로 된 비문이므로, 분석 편의를 위해 각 줄을 별도로 분리해, array로 넣는다. 

base_text = df[df.id==1]['reading'].values[0].split('\n')  # 줄바꿈(\n)을 바로 구분자로 나눈다. 

# 결과 확인 
print(len(base_text)) # 길이 12짜리 array로 쪼개졌는지 확인 
base_text
````

````python
# 같은 방식으로 비교할 연구별 비문 판독도 변수로 뽑아본다. 'comp_text'

id = 12  # 임의의 id 를 넣어본다. 

comp_text = df[df.id==id]['reading'].values[0].split('\n')

# 이 결과가 엑셀파일 'reading_rawdata' 시트에서 'reading' 컬럼의 값과 같은지 확인해본다. 
# id=12 는 '노중국' 연구. 시트에서 확인 결과 일치함을 알 수 있다.
comp_text 
````

````python
# 자세히보면 경주문화재연구소의 기준 판독에는 '/'와 공백이 없는데,
# 연구자들의 판독은 이런 기호가 있다. 
# 따라서 comp_text 변수에서 구분자(/)와 공백 삭제하자.

comp_text = [x.replace('/','').replace(' ','').strip() for x in comp_text]
````

````python
# 깔끔하게 글자들만 남은 것을 확인할 수 있다.
comp_text
````

````python
# difflib : 문자열 비교 패키지 
# https://docs.python.org/ko/3.9/library/difflib.html

from difflib import SequenceMatcher
````

````python
# 어떻게 동작하는지 테스트용으로 돌려봄
#
# SequenceMatcher 함수에 원본텍스트, 비교텍스트를 인자로 넣은 뒤 
# 리턴오브젝트에서 필요한 함수를 불러 쓰는 방식 

s = SequenceMatcher(None, base_text[0], comp_text[0])
s.get_matching_blocks()
````

````python
# 두 문자열에서 다른 글자(한자)와 그 위치를 찾아 알려주는 코드 


# 확인하고 싶은 연구를 id를 바꿔가며 해 보자. 이 경우는 id=2 부터 id=33 까지 넣어볼 수 있다.
# 일부러 자동으로 돌려놓지 않고, id를 하나씩 바꿔가며 확인한 뒤, 
# 출력 결과를 보고 엑셀파일의 'reading' 항목에서 수정함. 
id = 21

col_name = 'reading'

# base_text : 기준 판독문 (id=1)
# id=1 은 경문연 2019 버전으로 판독(reading)의 기준으로 삼는다. 
base_text = df[df.id==1][col_name].values[0].split('\n')

# comp_text : 비교할 연구의 판독문 (id: 2~33 중 하나)
comp_text = df[df.id==id][col_name].values[0].split('\n')
comp_text = [x.replace('/','').replace(' ','').strip() for x in comp_text]  # 동일한 기준에서 비교할 수 있게 기준판독문에는 없는 '/'와 공백문자 제거한다.


# 결과 출력
# 사람이 확인하기 좋게 기준, 비교 판독문을 각각 출력한다. 
print(f'# {id}')
print(df[df.id==id][['researcher','published']].values[0]) # 이 id가 누구의 연구인지 확인할 수 있게 찍어주는 용도 
print('=========================================')
print('\n'.join(base_text))
print('-----------------------------------------')
print('\n'.join(comp_text))
print('-----------------------------------------')

# 실제 달라지는 부분 출력 
for idx, (b, c) in enumerate(zip(base_text, comp_text)): # 변수의 각 줄을 비교한다. 중성리비는 총 12번 돌게 된다. 
    line = idx + 1
    print(f'--- line {line} ----------------------') # 판독문에서 몇번째 줄의 글자인지 확인 
    s = SequenceMatcher(None, b,c, autojunk=False)
    
    # get_opcodes() 함수가 바뀐 글자와 위치를 tuple로 리턴하므로 목적에 가장 부합한다. 자세한 설명은 아래 링크로 
    # https://docs.python.org/ko/3.9/library/difflib.html#difflib.SequenceMatcher.get_opcodes
    for tag, i1, i2, j1, j2 in s.get_opcodes(): 
        if tag == 'equal': # 비교할 내용이 동일하면, 볼 필요가 없으므로 그냥 넘긴다.
            continue
        print('{:7}   base[{}:{}] --> comp[{}:{}] {!r:>5} --> {!r}'.format(
            tag, i1, i2, j1, j2, b[i1:i2], c[j1:j2]))
    
    # 간혹 줄마다 글자 수가 서로 다른 경우가 있다. 이 부분에 오독이 없도록 확인용으로 출력 
    if len(b) != len(c):
        print(f'lendiff {len(b)} vs. {len(c)}')
````

````python
### 결과 해석 예시

# 출력 
"""
--- line 5 ----------------------
replace   base[9:10] --> comp[9:10]   '夲' --> '本'
--- line 6 ----------------------
replace   base[0:1] --> comp[0:2]   '□' --> '△△'
lendiff 20 vs. 21
"""

# 해석
"""
line 5 => 5번째 줄에서 
'replace' => 변경이 있음
'base[9:10] => 원본(base_text)에서 10번째 글자인 '夲'이 비교(comp_text)문에서 10번째 글자인 '本'으로 되어 있다.

line 6 => 6번째 줄에서
'replace' => 변경이 있음
'base[0:1] => 원본(base_text)에서 1번째 글자인 '□'이 비교(comp_text)문에서 1번째로 있는 '△△'로 되어 있다. 
lendiff 20 vs. 21 => 원본은 6번째 줄의 글자수가 20인데, 비교문은 21개이다.  (이유는 바로 위에 있음. 원본은 '□'만 있는데, 비교문은 '△△'가 들어있기 때문)
"""
````

## 구두점 위치를 자동으로 추출하는 방법 


````python
# 구두점 (punctuation) 위치 추출 
#
# 모든 연구에서 고리점에 해당하는 부분만 '/'로 표시했으므로 
# 어떤 글자 뒤에 고리점이 있는지만 추출해보기로 한다. 

for id in range(3,33+1): # 엑셀파일에서 전체 연구 개수는 33개. 이 중 처음 2개(id=1,2)는 경주문화재연구소라서 비교할 필요가 없다. 
    print(id)
    print(df[df.id==id][['researcher','published']].values[0]) # 연구 정보 출력 
    
    col_name = 'reading_refined' # 정말 글자가 다른 것만 비교하기 위해 정제해 둔 컬럼 
    comp_text = df[df.id==id][col_name].values[0]

    # 비교를 위해 변수를 두 개 만들어 쓴다. comp_text_ch, comp_text_punc 
    # 한자만 있는 문자열 : comp_text_ch 
    # 구두점('/') 기호만 남겨둔 문자열 : comp_text_punc
    comp_text_ch = comp_text.replace('\n','').replace('/','').replace(' ','')
    comp_text_punc = comp_text.replace('\n','').replace(' ','')
    
    # 글자수 검증 : 203 이 되어야 한다. 가끔 202인 것도 있으니 참고용으로 출력 
    print(len(comp_text_ch))

    print(comp_text_punc) # 확인용으로 출력 
    
    idx = 0
    for c in comp_text_punc:
        if c == '/':
            print(idx, comp_text_ch[idx-1], '고리점')
        else:
            idx += 1
````

### 연구자별 비문 판독과 표점 정보를 아래와 같은 json 포맷으로 넣는다. 


```json

```

````python
# 멀티 컬럼 확인용
list(df_read.columns)[8:15]
````

````python
# '이우태' 연구에서 한자판독('character') 컬럼만 뽑아보자. 
# 빈 값은 경주문화재연구소의 기준 판독과 같은 것. 내용이 있는 것만 이우태 연구에서 다른 판독글자임 

df_read['이우태','character']
````

````python
studies = []

authors = {}
for author, feature in list(df_read.columns):
    if author in authors:
        authors[author].append(feature)
    else:
        authors[author] = [feature]

del authors['기준']

for author, v in authors.items():
    reading = {}
    punctuation = {}
    
    # 연구자별로 판독이 다른 한자와 훈/음 정보 추출 
    idx = 0
    for c,s,m,d in zip(list(df_read[author,'character']),
                       list(df_read[author,'sound']),
                       list(df_read[author,'meaning']),
                       list(df_read[author,'desc'])):
        idx += 1
        if c == '' and s == '' and m == '' and d == '':
            continue
        reading[idx] = {
            'character': c,
            'sound': s,
            'meaning': m,
            'desc': d
        }
    
    # 연구자별로 구두점 정보 추출
    idx = 0
    for p in list(df_read[author,'punctuation']):
        idx += 1
        if p == '':
            continue
            
        punctuation[idx] = {
            'type': p
        }
    
    # 최종 결과 dictionary로 만들기 
    studies.append({
        'articles' : [
            {
                'author': author,
                'doi': '',
                'title': '',
                'published': 2022,
                'publisher': '',
                'no': '',
            }
        ],
        'study': {
            'reading': reading,
            'punctuation': punctuation,
        }
    })
````

````python
# 결과
# 'studies' 변수는 dictionary 로 json이 아니다. 나중에 dictionary => json 으로 변환해서 파일로 저장할 예정임 

studies
````

````python
# 최종 결과. 하나의 json으로 묶기 위해서, 전체 dictionary를 만든다. 

full_data = {
    'monument': monument,
    'studies' : studies
}
````

````python
# json 파일 저장 

import json 

with open("silla-jungseong-ri-study.json", "w") as outfile:
    json.dump(full_data, outfile, ensure_ascii=False, indent=2) # 보기 편하게 띄어쓰기 2로 설정 
````
